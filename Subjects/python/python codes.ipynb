```python
data = {
"Student": [
"Aman",
"Riya",
"Karan",
"Simran",
"Rahul",
"Neha",
"Arjun",
"Pooja"
],
"Department": [
"CSE",
"CSE",
"ECE",
"ECE",
"CSE",
"ECE",
"CSE",
"ECE"
],
"Math_Marks": [
78,
92,
67,
85,
74,
88,
81,
70
],
"Python_Marks": [
82,
95,
72,
80,
69,
91,
86,
75
],
"Attendance": [
85,
96,
75,
90,
70,
94,
88,
78
],
"Result": [
"Pass",
"Pass",
"Pass",
"Pass",
"Pass",
"Pass",
"Pass",
"Pass"
"Pass"
]
}
df = pd.DataFrame(data)
print(df)

```

```python
sns.barplot(
data=df,
x="Student",
y="Python_Marks"
)
plt.title("Python Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.barplot(
data=df,
x="Student",
y="Python_Marks"
)
plt.title("Python Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(
data=df,
x="Python_Marks",
kde=True
)
plt.title("Distribution of Python Marks")
plt.xlabel("Python Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(
data=df,
x="Attendance",
y="Python_Marks",
hue="Department"
)
plt.title("Attendance versus Python Marks")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```
